```python
from datetime import datetime, timedelta
df.assign(trx = np.ones(df.shape[0])).set_index(df['Date']).resample(timedelta(days=1)).count().trx.plot(title='Daily Transactions')
```

```python
df.assign(trx = np.ones(df.shape[0])).set_index(df['Date']).resample(timedelta(days=7)).count().trx.plot(title='Weekly Transactions')
```

```python
goods = df[df['Fraud']==0]
bads = df[df['Fraud']==1]
ngoods = len(goods)
nbads = len(bads)
goods_series = goods.assign(apps = np.ones(ngoods)).set_index(goods['Date']).resample(timedelta(days = 7)).count().apps
norm_goods_series = goods_series / ngoods
norm_goods_series.plot(title = 'Weekly Transactions', color = 'green')
bads_series = bads.assign(apps = np.ones(nbads)).set_index(bads['Date']).resample(timedelta(days = 7)).count().apps
norm_bads_series = bads_series / nbads
norm_bads_series.plot(color = 'red')
```

```python
df.assign(trx = np.ones(df.shape[0])).set_index(df['Date']).resample(timedelta(days=30)).count().trx.plot(title='Approximately Monthly Transactions')
```
