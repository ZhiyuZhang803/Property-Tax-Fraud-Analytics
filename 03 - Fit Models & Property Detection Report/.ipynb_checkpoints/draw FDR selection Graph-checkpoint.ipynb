```python
from datetime import datetime
start_time = datetime.now()

# Libraries to load
# !pip install lightgbm

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
import lightgbm as lgb
import numpy as np
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")

print('LOAD DURATION: ', datetime.now() - start_time) # load time about 30 seconds
```

## Preparation for Model Fitting

```python
%%time
vars = pd.read_csv('vars_final.csv')
print(vars.shape)
vars.head()
```

```python
final_vars = pd.read_csv('final_vars_list.csv')
final_vars
```

```python
numvars = min(15,len(final_vars))
final_vars_list = ['Recnum','Fraud']
for i in range(numvars):
    final_vars_list.append(final_vars.iloc[i]['variable name'])
    
final_vars_list
```

```python
vars = vars.filter(final_vars_list,axis=1)
vars.head()
```

```python
# use this to cap variables. For some problems it helps
Clip = 10
```

```python
vars.rename(columns={'fraud_label':'Fraud'},inplace=True)
```

```python
vars['Fraud'].sum()
```

```python
vars.head(10)
```

```python
vars.shape
```

```python
vars.describe()
```

```python
record_save = vars['Recnum']
Y_save = pd.DataFrame(vars.loc[:,'Fraud'])
Y_save.head()
```

## Scale and truncate field values

```python
X_no_scaling = vars.drop(columns = ['Recnum','Fraud'])
X_no_scaling.describe()
```

```python
X = (X_no_scaling - X_no_scaling.mean()) / X_no_scaling.std()
```

```python
# push in any outlier values
cols = X.columns
X.loc[:,cols] = X[cols].clip(upper=Clip)
X.loc[:,cols] = X[cols].clip(lower=-1*Clip)
# X = (X_no_scaling - X_no_scaling.mean()) / X_no_scaling.std()
X.describe()
```

```python
# separate data into modeling (traintest) and out of time
oot_recnum=84300
X_trntst = X[0:oot_recnum]
Y_trntst = Y_save[0:oot_recnum]
X_oot = X[oot_recnum:]
Y_oot = Y_save[oot_recnum:]
```

```python
niter = 0
nitermax = 10
FDR3 = pd.DataFrame(np.zeros((nitermax,3)), columns=('trn', 'tst', 'oot'))
X_oot_orig = X_oot.copy()
pd.options.mode.chained_assignment = None  # default='warn'
```

## Fit Best Model

```python
%%time
# LightGB

train_value = []
test_value = []
oot_value = []
X_trntst_new = X_trntst.iloc[:,:5]
X_oot_new = X_oot_orig.iloc[:,:5]

X_trn, X_tst, Y_trn, Y_tst = train_test_split(X_trntst_new, Y_trntst, test_size = .3,random_state=10)

model = lgb.LGBMClassifier(n_estimators=100,max_depth=4,learning_rate=0.01,subsample=0.8)

X_oot = X_oot_orig.copy()
X_trn_save = X_trn.copy()
Y_trn_save = Y_trn.copy()

model.fit(X_trn, Y_trn.values.ravel())   

predictions = model.predict_proba(X_trn_save)[:,1]
X_trn['predicted'] = predictions
X_trn['Fraud'] = Y_trn_save['Fraud']

predictions = model.predict_proba(X_tst)[:,1]
X_tst['predicted']=predictions
X_tst['Fraud'] = Y_tst['Fraud']

predictions = model.predict_proba(X_oot_new)[:,1]
X_oot['predicted']=predictions
X_oot['Fraud'] = Y_oot['Fraud']
temp = X_oot.sort_values('predicted',ascending=False)
```

```python
temp.head(5)
```

## Add the columns that are used to draw a graph and draw a graph

```python
amount1 = 0
fraud_saving = []
for i in temp['Fraud']:
    if i == 0:
        fraud_saving.append(amount1)
    else:
        amount1 = amount1+2000
        fraud_saving.append(amount1)
```

```python
temp['fraud_saving'] = fraud_saving
temp.head()
```

```python
amount2 = 0
fraud_cost = []
for i in temp['Fraud']:
    if i ==0:
        amount2 = amount2+50
        fraud_cost.append(amount2)
    else:
        fraud_cost.append(amount2)
```

```python
temp['fraud_cost'] = fraud_cost
temp.head()
```

```python
label = [i+1 for i in range(temp.shape[0])]
temp['label'] = label
temp.head(5)
```

```python
temp['diff'] = temp['fraud_saving']-temp['fraud_cost']
```

```python
import matplotlib.pyplot as plt
plt.plot(temp['label'],temp['fraud_saving'])
plt.plot(temp['label'],temp['fraud_cost'])
plt.plot(temp['label'],temp['diff'])
plt.show()
```

```python
final_saving = []
for m,n in zip(temp['fraud_saving'],temp['label']):
    if n%121 !=0:
        continue
    else:
        final_saving.append(m)
final_saving
```

```python
final_cost = []
for m,n in zip(temp['fraud_cost'],temp['label']):
    if n%121 !=0:
        continue
    else:
        final_cost.append(m)
final_cost
```

```python
final_plot_use = pd.DataFrame(data={'Fraud_saving':final_saving,'Fraud_cost':final_cost})
```

```python
final_plot_use['Total_Saving'] = final_plot_use['Fraud_saving'] - final_plot_use['Fraud_cost']
```

```python
label2 = [i+1 for i in range(final_plot_use.shape[0])]
final_plot_use['Label'] = label2
```

```python
final_plot_use.head(5)
```

```python
final_plot_use = final_plot_use[final_plot_use['Label']<=50]
```

```python
plt.rcParams.update({'figure.figsize':(12,8)})
plt.rcParams.update({'font.size':20})
```

```python
plt.plot(final_plot_use['Fraud_saving'],label='Fraud Savings')
plt.plot(final_plot_use['Fraud_cost'],label='Lost Sales')
plt.plot(final_plot_use['Total_Saving'],label='Overall Savings')
plt.xlabel("OOT Population % Bins")
plt.title("Choice of FDR")
plt.ylabel("Amount of Dollars ($)")
plt.vlines(ymin=0,ymax=330000,x=3,colors='black')
plt.legend(loc='center right',prop={'size': 15})
plt.margins(0.0)
plt.show()
```
