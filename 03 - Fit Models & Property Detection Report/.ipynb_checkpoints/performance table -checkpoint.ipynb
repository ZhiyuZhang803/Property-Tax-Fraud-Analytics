```python
from datetime import datetime
start_time = datetime.now()

# Libraries to load
# !pip install lightgbm

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
import lightgbm as lgb
import numpy as np
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")

print('LOAD DURATION: ', datetime.now() - start_time) # load time about 30 seconds
```

## Preparation for Model Fitting

```python
%%time
vars = pd.read_csv('vars_final.csv')
print(vars.shape)
vars.head()
```

```python
final_vars = pd.read_csv('final_vars_list.csv')
final_vars
```

```python
numvars = min(15,len(final_vars))
final_vars_list = ['Recnum','Fraud']
for i in range(numvars):
    final_vars_list.append(final_vars.iloc[i]['variable name'])
    
final_vars_list
```

```python
vars = vars.filter(final_vars_list,axis=1)
vars.head()
```

```python
# use this to cap variables. For some problems it helps
Clip = 10
```

```python
vars.rename(columns={'fraud_label':'Fraud'},inplace=True)
```

```python
vars['Fraud'].sum()
```

```python
vars.head(10)
```

```python
vars.shape
```

```python
vars.describe()
```

```python
record_save = vars['Recnum']
Y_save = pd.DataFrame(vars.loc[:,'Fraud'])
Y_save.head()
```

## Scale and truncate field values

```python
X_no_scaling = vars.drop(columns = ['Recnum','Fraud'])
X_no_scaling.describe()
```

```python
X = (X_no_scaling - X_no_scaling.mean()) / X_no_scaling.std()
```

```python
# push in any outlier values
cols = X.columns
X.loc[:,cols] = X[cols].clip(upper=Clip)
X.loc[:,cols] = X[cols].clip(lower=-1*Clip)
# X = (X_no_scaling - X_no_scaling.mean()) / X_no_scaling.std()
X.describe()
```

```python
# separate data into modeling (traintest) and out of time
oot_recnum=84300
X_trntst = X[0:oot_recnum]
Y_trntst = Y_save[0:oot_recnum]
X_oot = X[oot_recnum:]
Y_oot = Y_save[oot_recnum:]
```

```python
niter = 0
nitermax = 10
FDR3 = pd.DataFrame(np.zeros((nitermax,3)), columns=('trn', 'tst', 'oot'))
X_oot_orig = X_oot.copy()
pd.options.mode.chained_assignment = None  # default='warn'
```

## Fit Best Models

```python
%%time
# LightGB

train_value = []
test_value = []
oot_value = []
X_trntst_new = X_trntst.iloc[:,:5]
X_oot_new = X_oot_orig.iloc[:,:5]

X_trn, X_tst, Y_trn, Y_tst = train_test_split(X_trntst_new, Y_trntst, test_size = .3,random_state=10)

model = lgb.LGBMClassifier(n_estimators=100,max_depth=4,learning_rate=0.01,subsample=0.8)

X_oot = X_oot_orig.copy()
X_trn_save = X_trn.copy()
Y_trn_save = Y_trn.copy()

model.fit(X_trn, Y_trn.values.ravel())   

predictions = model.predict_proba(X_trn_save)[:,1]
X_trn['predicted'] = predictions
X_trn['Fraud'] = Y_trn_save['Fraud']

predictions = model.predict_proba(X_tst)[:,1]
X_tst['predicted']=predictions
X_tst['Fraud'] = Y_tst['Fraud']

predictions = model.predict_proba(X_oot_new)[:,1]
X_oot['predicted']=predictions
X_oot['Fraud'] = Y_oot['Fraud']
```

```python
import math
```

## Draw Train Performance Table

```python
NumOfRows_oneBin = math.ceil(int(round(X_trn.shape[0]))/100)
NumOfRows_oneBin_1 = math.ceil(int(round(X_tst.shape[0]))/100)
NumOfRows_oneBin_2 = math.ceil(int(round(X_oot.shape[0]))/100)

# for train
binNum = []
NumRecords = []
NumGoods = []
NumBads = []
PercentGoods = []
PercentBads = []
TotalRecords = []
CumGoods = []
CumBads = []
CumPercentGood = []
CumPercentBad = []
KS = []
FPR = []
total_Goods = X_trn[X_trn['Fraud']==0].shape[0]
total_Bads = X_trn[X_trn['Fraud']==1].shape[0]
for i in range(1,21):
    new_df = X_trn.sort_values('predicted',ascending=False)[(i-1)*NumOfRows_oneBin:i*NumOfRows_oneBin]
    binNum.append(i)
    NumRecords.append(NumOfRows_oneBin)
    NumGoods.append(new_df[new_df['Fraud']==0].shape[0])
    NumBads.append(new_df[new_df['Fraud']==1].shape[0])
    PercentGoods.append(round(NumGoods[-1]*100/NumOfRows_oneBin,2))
    PercentBads.append(100-PercentGoods[-1])
    TotalRecords.append(i*NumOfRows_oneBin)
    CumGoods.append(sum(NumGoods))
    CumBads.append(sum(NumBads))
    CumPercentGood.append(round(CumGoods[-1]*100/total_Goods,2))
    CumPercentBad.append(round(CumBads[-1]*100/total_Bads,2))
    KS.append(CumPercentBad[-1]-CumPercentGood[-1])
    FPR.append(round(CumGoods[-1]/CumBads[-1],2))

good = pd.DataFrame(data={'Pop':binNum,'NumberRecords':NumRecords,'Good':NumGoods,'Bad':NumBads,'PercentGood':PercentGoods,'PercentBad':PercentBads,'TotalRecodes':TotalRecords,'CumGood':CumGoods,'Cumbad':CumBads,'CumPercentGood':CumPercentGood,'CumPercentBad':CumPercentBad,'KS':KS,'FPR':FPR})
good
```

```python
good.to_excel('good.xlsx')
```
